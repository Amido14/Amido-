```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('user_behavior_dataset (1).csv')
```

```python
print(df.head())
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
print("\
Number of duplicate rows:", df.duplicated().sum())
```

```python
print("\
Unique values in 'Device Model':", df['Device Model'].nunique())
print("Unique values in 'Operating System':", df['Operating System'].nunique())
print("Unique values in 'Gender':", df['Gender'].unique())
print("Unique values in 'User Behavior Class':", df['User Behavior Class'].unique())
```

```python
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

sns.countplot(data=df, x='User Behavior Class', hue='Gender', ax=ax, palette=['#2563EB', '#24EB84'])

ax.set_title('Gender Distribution by User Behavior Class', pad=15, fontsize=20, color='#171717', weight='semibold')
ax.set_xlabel('User Behavior Class', labelpad=10, fontsize=16, color='#171717', weight='medium')
ax.set_ylabel('Count', labelpad=10, fontsize=16, color='#171717', weight='medium')
ax.grid(axis='y', color='#F3F4F6')
ax.set_axisbelow(True)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#171717')
ax.spines['bottom'].set_color('#171717')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

sns.boxplot(data=df, x='User Behavior Class', y='Age', ax=ax, palette='muted')

ax.set_title('Age Distribution by User Behavior Class', pad=15, fontsize=20, color='#171717', weight='semibold')
ax.set_xlabel('User Behavior Class', labelpad=10, fontsize=16, color='#171717', weight='medium')
ax.set_ylabel('Age', labelpad=10, fontsize=16, color='#171717', weight='medium')
ax.grid(axis='y', color='#F3F4F6')
ax.set_axisbelow(True)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#171717')
ax.spines['bottom'].set_color('#171717')

plt.show()
```
